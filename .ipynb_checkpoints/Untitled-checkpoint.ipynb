```python
import pandas as pd

# URL do arquivo CSV com os dados
url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'

# Carregar os dados para um DataFrame chamado 'df'
df = pd.read_csv(url)

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.groupby('Sex')['Survived'].mean()
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.show()
```

```python
df.groupby('Sex')['Age'].mean()
```

```python
mulheres_sobreviventes = df[(df['Sex'] == 'female') & (df['Survived'] == 1)]
media = mulheres_sobreviventes['Age'].mean()
```

```python
print(media)
```

```python
sns.countplot(x='Survived', hue='Sex', data=df)
plt.show()
```

```python
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.show()
```

```python
df.groupby('Pclass')['Survived'].mean()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.show()
```

```python
df['FamilySize'] = df['SibSp'] + df['Parch'] + 1
df['IsAlone'] = df['FamilySize'] == 1

```

```python
df.head()
```

```python
df.groupby('IsAlone')['Survived'].mean()
```

```python
homem_adulto_terceira = df[(df['Age'] > 18) & (df['Sex'] == 'male') & (df['Pclass'] == 3)]
```

```python
taxa = homem_adulto_terceira['Survived'].mean()
print(taxa)
```

```python
df['Title'] = df['Name'].str.extract(r'([A-Za-z]+)\.', expand=False)
```

```python
print(df['Title'].value_counts())
```

```python
corr_matrix = df.corr(numeric_only=True)
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
```

```python
import pandas as pd

caminho_arquivo = 'zabbix_memorias.txt'

try:
    df = pd.read_csv(
        caminho_arquivo,
        sep='\s+',
        header=None,
        names=['data', 'hora', 'clock', 'valor'],
        skiprows=1
    )

    df['timestamp_str'] = df['data'] + ' ' + df['hora']

    df['timestamp'] = pd.to_datetime(df['timestamp_str'], format='%d-%m-%Y %H:%M:%S')

    df.set_index('timestamp', inplace=True)

    df.drop(['data', 'hora', 'clock', 'timestamp_str'], axis=1, inplace=True)

    print("Dados carregados e formatados\n")
    print(df.head())
    #print(df.info())

    print(df['valor'].idxmax())

except FileNotFoundError:
    print(f"Erro")
    print("e")
```
